```python
def meanSquareError(height,width,I1,I2):
    sq=[]
    i=0
    while i<height:
        j=0
        while j<width:
            sq.append(np.square(I1[i,j]-I2[i,j]))
            j+=1
        i+=1

    #print(len(sq))
    #print('Square',np.square(sq))

    k=np.sum(sq)/(height*width)
    return(k)
```

```python
print('Manual MSE : ',meanSquareError(height,width,I1,I2)) #manually computed mean square error
```
