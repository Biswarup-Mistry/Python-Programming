```python
def mseplotting(img,item):
    i=20
    x=[]
    y=[]
    d=0.0
    while(i>0 and d<=1.0):
        noise_img=get_Noise(img,d) #getting noisy image

        mse=np.square(np.subtract(img,noise_img)).mean() #calculating MSE of Original image and Image with Noise
        #print('MSE : ',mse)
        x.append(d)
        y.append(mse)
        d=d+0.05 #d runs for 20 times
        
        
        
        
        
        #f = open('MSE(s)ofImages.csv', "a") 
        #f.write(str(mse)+'\n')
        #f.close()
        i=i-1

    plt.title('MSE VS Density of Noise of Image : '+item)
    plt.xlabel("Density")
    plt.ylabel("MSE")
    #plt.scatter(x, y)

    plt.plot(x, y, marker='o')
    #plt.twinx()
    plt.show()  #need to be turned on-------
    
    #f = open('MSE(s)ofImages.csv', "a") 
    #f.write(str(y))
    #f.write('\n')
    #f.close()
    
    
      
    
    
    
    
    df= pd.read_csv('MSE(s)ofImages.csv')                           # Read CSV file
    df = pd.DataFrame(columns=['density/Item'])
    #df.insert(loc=item,column=item value=y)
    #df.to_csv("MSE(s)ofImages.csv")
    
    df.insert(loc=len(df.columns),                       # Where to insert
             column=item,                        # Name of new column
             value=y)
    
    
    #df = df.assign(item = y) # New column
                       #new_4 = mtcars.mpg ï¼ž 30) # Another new column
    df.to_csv("MSE(s)ofImages.csv")
    
```
