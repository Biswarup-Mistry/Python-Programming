```python
# importing cv2
import cv2
import matplotlib.pyplot as plt
```

```python
path=r'C:\Users\Biswarup Mistry\Desktop\4.1.05.png'
img=cv2.imread(path)
```

```python
Lena=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(Lena,cmap='gray')
```

```python
i=0

while i<8:
    outputpath='C://Users//Biswarup Mistry//Desktop//bp//BP'
    #bi=str('b'+str(i))
    bi=(Lena>>i) & 1
    b=bi*255
    #print()
    outputpath=str(outputpath+str(i)+'.png')
    cv2.imwrite(outputpath,b)
    #plt.imshow(b,cmap='gray')
    i+=1
```
