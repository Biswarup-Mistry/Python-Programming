```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('C:\\Users\\Biswarup Mistry\\Desktop\\Accounting of Study1.csv')
#data
```

```python
def dataExtractingWithOne() :# ----------------------- [Data Extracing (Begin) only studing days with 1]
    k,m = data['Studyied (0/1) ?'], data['Hours']
    #print('K :\n',k)
    #m
    y=[]
    x=[]
    i=0
    count=0
    while i<len(k):
        if k[i]==1:
            y.append(m[i])
            count+=1
        i+=1

    i=0
    while i<count:
        x.append(i)
        i+=1
    plt.title('Days(x) vs Hours(y) [Sitting Days Only]')
    plt.plot(x,y, marker = 'o') #only studing days with 1
    plt.grid()
    plt.show() 
    print('# of Studing Days : ',count,'/',len(m))
    
def dataExtractingNormal():# ----------------------- [Normal]
    i=0
    x=[]
    while i<len(m):
        x.append(i)
        i+=1
    plt.title('Days(x) vs Hours(y) [Normal]')
    plt.plot(x,m,  marker = 'o') #every day included
    plt.grid()
    plt.show()
    i=0
    count=0
    while i<len(k):
        if k[i]==1:
            count+=1
        i+=1
    print('# of Studing Days : ',count,'/',len(m))
    
```

```python
dataExtractingNormal() #includes no study days too
```

```python
dataExtractingWithOne() #excludes no study days
```
