```python
def Truth():
    m=data['Hours']
    #print('m= ', m)
    x=[]
    i=0
    x=m.dropna()
    #print('x=',x)
    #print('len(x)=',len(x))
    #len(y)=len(x)
    y=[None]*len(x)
    while i<len(x):
        y[i]=i
        i+=1
    #print('y=',y)
    plt.plot(y,x)
    plt.show()
    i=0
    count=0
    max=0
    while i<len(x):
        if x[i]!=0:
            count+=1
            if max<x[i]:
                max=x[i]
        i+=1
    print('(i). Total Day of Studing : ',count,'/',len(x))
    p=count/len(x)
    print('(ii). Probability of Success : ', p*100,' %')
    print('(iii). Highest hour of study : ',max,' Hour(s)')
```
