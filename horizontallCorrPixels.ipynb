```python
def pearson_corr(fi,fj):
    #a=[]
    #b=[]
    #a=pixel_Matrix[:, fi]
    #b=pixel_Matrix[:, fj]
    #x=np.array(a.flatten())
    #y=np.array(b.flatten())
    x=[]
    y=[]
    x=pixel_Matrix[:, fi]
    y=pixel_Matrix[:, fj]
    
    #print(x)
    #print(y)
    k_x=[]
    k_y=[]
    k_x_sq=[]
    k_y_sq=[]
    l=[]
    
    #print('len(x)',len(x))
    #print('len(y)',len(y))
    sum_x=np.sum(x)
    sum_y=np.sum(y)
    x_bar=sum_x/len(x)
    y_bar=sum_y/len(y)
    for i in x:
        k_x.append(i-x_bar)
    for i in y:
        k_y.append(i-y_bar)
    for i in k_x:
        k_x_sq.append(i*i)
    for i in k_y:
        k_y_sq.append(i*i)
    sum_k_x_sq=np.sum(k_x_sq)
    std_dv_x=sum_k_x_sq/len(x)
    sum_k_y_sq=np.sum(k_y_sq)
    std_dv_y=sum_k_y_sq/len(y)
    i=0
    while i<len(x):
        l.append(k_x[i]*k_y[i])
        i+=1
    sum_of_l=np.sum(l)
    cov_x_y=sum_of_l/len(x)
    #pears_corr=cov_x_y/((math.sqrt(std_dv_x))*(math.sqrt(std_dv_y)))
    pears_corr=float("{:.2f}".format(cov_x_y/((math.sqrt(std_dv_x))*(math.sqrt(std_dv_y)))))
    return pears_corr
    #del x
    #del y
    #del k_x
    #del k_y
    #del k_x_sq
    #del k_y_sq
    #del l
    
```
