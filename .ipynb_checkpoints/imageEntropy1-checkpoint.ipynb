```python
def EntropyOfImage():
    a=[] #for taking every pixels into a single array
    d=[] #for putting entropy of every pixels
    c=[] #for putting every pixels' probability

    i=j=0
    while i<height: # taking pixels' valus to an single array
        j=0
        while j<width:
            a.append(I[i,j])
            j+=1
        i+=1


    i=0
    max=a[i]
    while i+1<len(a): # finding the max
        if(max<a[i+1]):
            max=a[i+1]
        i+=1

    b=np.asarray(np.arange(max+1).reshape(max+1)) # create an array of size 'max'
    i=0
    while i<len(b): # like counting sort, putting 0 into every index of array 'b'
        b[i]=0
        i+=1

    i=0
    while i<len(a): # at a[i]th index of array 'b', counting is taking place w.r.t each value of array 'a'
        b[a[i]]+=1
        i+=1
    i=0
    sum=np.sum(b) # for getting the probability
    while i<len(a):
        p=b[a[i]]/sum # computing probability for each value
        c.append(p)
        i+=1
    i=0
    while i<len(c):
        d.append(c[i]*np.log10(c[i])) #getting entropy for each probability
        i+=1
    e=-np.sum(d) #sum of the array 'd'
    e=np.log2(e)
    return(e) #returning the entropy 'e' of the Image 
```
