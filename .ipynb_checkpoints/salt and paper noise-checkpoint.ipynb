```python
#from SSIM_PIL import compare_ssim
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import random
import pandas as pd


import skimage
from skimage.io import imread
import os

iteration=20
```

```python
path='C://Users//Biswarup Mistry//Desktop//Misc//1.tiff'  #defining the image directory path
img = imread(path)
    
plt.title('Original Image : 1.tiff ')
plt.imshow(img,cmap=plt.cm.gray)
plt.show()
```

```python
#return skimage.util.random_noise(img, mode='s&p', amount=d)
d=0.1
paper=d
salt=1-d
row, col=img.shape
g=np.matrix(img).reshape(row, col)
for i in range(row):
    for j in range(col):
        ran=np.random.random()
        if ran<paper:
            g[i,j]=0
        elif ran>salt:
            g[i,j]=255
        else:
            g[i,j]=img[i,j]
plt.imshow(g,cmap=plt.cm.gray)
plt.show()
```
